```python
import subprocess
try:
    x = subprocess.call(['sh', './status.sh'])
    #print("error code", x.returncode, x.stdout, x.stderr)
    x.check_call()

except subprocess.CalledProcessError as grepexc:                                                                                                   
    print("error code", grepexc.returncode, grepexc.output)
```
