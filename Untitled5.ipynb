```
edges = [[0,1],[0,2],[1,3],[2,4],[3,5],[4,5],
         [0,6],[1,9],[2,10],[2,14],[3,13],[3,17],[4,18],[5,21],
         [6,7],[6,10],[7,8],[8,9],[9,13],
         [10,11],[11,12],[11,15],[12,13],[12,16],
         [14,15],[14,18],[15,16],[16,17],[17,21],
         [18,19],[19,20],[20,21]]

n = 22

deg = np.array(np.zeros(n,dtype=int))

vert = {}

for i in edges:
  deg[i[0]] += 1
  deg[i[1]] += 1
  
  if i[0] not in vert:
    vert[i[0]] = {i[1]}
  else:
    if i[1] not in vert[i[0]]:
      vert[i[0]].add(i[1])
  
  if i[1] not in vert:
    vert[i[1]] = {i[0]}
  else:
    if i[0] not in vert[i[1]]:
      vert[i[1]].add(i[0])

A = np.array(np.zeros((n,n)))

for i in range(22):
  for j in range(22):
    if i==j:
      A[i,i] = 1
    else:
      if i in vert[j]:
        A[j,i] = -1/deg[j]
      else:
        A[j,i] = 0
        
print(A)        
    

  
```
