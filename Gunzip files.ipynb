# **Unzip .GZ file (not tar.gz)**


```
import gzip
import shutil
with gzip.open('../content/listings.csv.gz', 'rb') as f_in:
    with open('../content/listings.csv', 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)

```

#**Unpack tar file**

```
import tarfile
with tarfile.open('../content/cifar-10-python.tar.gz') as tar:
    tar.extractall(path='../content/')
```

#**Unpicle the file (if the file is binary format then read it from binary)**

```
import pickle
def unpickle(file, m='readfile'):
  with open(file, 'rb') as fo:
    dict = pickle.load(fo, encoding='bytes')
    if m=='creatfile':
      with open(file+'1','w') as fi:
        fi.write(str(dict))
  return dict

x = unpickle('../content/cifar-10-batches-py/data_batch_1', 'creatfile')

```
